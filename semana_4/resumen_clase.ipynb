## Ontología

Es una representación y estructurada del conocimiento sobre un dominio específico. Define los conceptos (clases), las relaciones entre ellos y las reglas que describen cómo interactúan. Las ontologías se usan para organizar, compartir y reutilizar información de manera que las máquinas puedan entenderla y procesarla, facilitando tareas como la integración de datos, la búsqueda semántica y la inteligencia artificial.

### Semantica

Semántica en informática se refiere al significado o interpretación de los datos, es decir, cómo se entienden y relacionan los conceptos dentro de un contexto específico.

Búsqueda semántica es una técnica que utiliza el significado de las palabras y las relaciones entre conceptos para mejorar los resultados de búsqueda. A diferencia de la búsqueda tradicional (que solo compara palabras clave), la búsqueda semántica entiende la intención del usuario y el contexto, ofreciendo resultados más relevantes y precisos.

### Web semantica

La Web Semántica es una extensión de la web tradicional cuyo objetivo es hacer que la información en internet sea comprensible no solo para las personas, sino también para las máquinas. Utiliza ontologías, metadatos y estándares como RDF (Resource Description Framework) y OWL (Web Ontology Language) para describir el significado y las relaciones entre los datos.

Está basada en la idea de proporcionar en la Web datos  definidos y enlazados, permitiendo que 
aplicaciones heterogéneas localicen, integren, razonen y reutilicen la información presente en la Web 

Esto permite que los sistemas automaticen tareas como la integración, búsqueda y análisis de información, facilitando la interoperabilidad entre diferentes aplicaciones y mejorando la precisión de los resultados. En la Web Semántica, los datos están conectados y estructurados de manera que los programas pueden interpretarlos y utilizarlos de forma inteligente.

### RDF

RDF (Resource Description Framework) es un estándar para describir y modelar información en la Web Semántica. Permite representar datos mediante tripletas: sujeto, predicado y objeto, lo que facilita expresar relaciones entre recursos de manera estructurada y comprensible tanto para humanos como para máquinas. 

Para usar RDF (Resource Description Framework), debes representar la información en forma de tripletas: sujeto, predicado y objeto. Puedes escribir RDF en varios formatos, como Turtle, RDF/XML o JSON-LD. Además, existen herramientas y librerías para crear y consultar datos RDF.


Estas tripletas forman afirmaciones del tipo:

Sujeto: la entidad o recurso del que se habla.

Predicado: la propiedad o relación que se afirma.

Objeto: el valor de la propiedad o el recurso relacionado.

En RDF, la URI (Uniform Resource Identifier) es el identificador único de cada nodo o recurso. Permite distinguir de manera global y sin ambigüedades cada entidad, propiedad o relación en los datos, asegurando que no haya confusión entre recursos aunque se usen en diferentes contextos o aplicaciones.

En ontologías, RDF (Resource Description Framework) es el modelo base para representar conocimiento mediante tripletas y Turtle es un formato que se puede usar 

## RDF Schema (RDFS)
 es una extensión de RDF que permite definir vocabularios, clases y propiedades para estructurar datos de manera más rica. RDFS introduce conceptos como:

rdfs:Class: para definir clases (tipos de recursos).

rdfs:subClassOf: para establecer jerarquías entre clases.

rdfs:Property: para definir propiedades.

rdfs:subPropertyOf: para jerarquías de propiedades.

rdfs:domain y rdfs:range: para especificar a qué clase pertenece una propiedad y qué tipo de valor puede tener.
