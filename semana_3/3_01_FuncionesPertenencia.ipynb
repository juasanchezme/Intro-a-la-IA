# SciKit-Fuzzy: Funciones de Pertenencia

En SciKit-Fuzzy, puedes implementar las siete funciones de pertenencia ms comnmente usadas en la lgica difusa. Estas funciones son Triangular, Trapezoidal, Gaussiana, Sigmoidal, Bell, S, Z.

```
!pip install scikit-fuzzy
```

```
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

**1. Funcin de Pertenencia Triangular (trimf)**

La funcin triangular se define por tres puntos: el inicio, el pico y el final.

```
x = np.linspace(0, 40, 1000)
mf_triangular = fuzz.trimf(x, [10, 20, 30])

plt.figure()
plt.plot(x, mf_triangular, 'b', linewidth=1.5, label='Triangular')
plt.title('Funcin de Pertenencia Triangular')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**2. Funcin de Pertenencia Trapezoidal (trapmf)**

La funcin trapezoidal se define por cuatro puntos: el inicio de la rampa ascendente, el inicio de la meseta, el final de la meseta y el inicio de la rampa descendente.

```
mf_trapezoidal = fuzz.trapmf(x, [10, 15, 25, 30])

plt.figure()
plt.plot(x, mf_trapezoidal, 'r', linewidth=1.5, label='Trapezoidal')
plt.title('Funcin de Pertenencia Trapezoidal')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**3. Funcin de Pertenencia Gaussiana**

La Gaussiana se caracteriza por su centro y desviacin estndar.

```
mf_gaussiana = fuzz.gaussmf(x, 20, 5)

plt.figure()
plt.plot(x, mf_gaussiana, 'g', linewidth=1.5, label='Gaussiana')
plt.title('Funcin de Pertenencia Gaussiana')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**4. Funcin de Pertenencia Sigmoidal**

La Sigmoidal se define por su pendiente (0.1) y el valor en el centro de la curva S (20).

```
mf_sigmoidal = fuzz.sigmf(x, 20, 1)

plt.figure()
plt.plot(x, mf_sigmoidal, 'k', linewidth=1.5, label='Sigmoidal')
plt.title('Funcin de Pertenencia Sigmoidal')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**5. Funcin de Pertenencia Bell (gbellmf)**

La funcin de Bell se define mediante tres parmetros: a, b, y c, donde a es el ancho, b es la pendiente y c es el centro de la curva.

```
# Definir el universo de discurso
x = np.linspace(-10, 10, 100)

# Parmetros para la funcin Bell
a = 2
b = 3
c = 0

# Generar los valores de membresa
y = fuzz.gbellmf(x, a, b, c)

# Graficar
plt.figure()
plt.plot(x, y, 'b', linewidth=1.5, label='Bell')
plt.title('Funcin de Pertenencia Bell')
plt.legend()
plt.show()
```

**6. Funcin de Pertenencia S (smf)**

La funcin S es til para modelar transiciones suaves y se define por dos parmetros: a y c, que son los puntos de inflexin de la curva.

```
# Parmetros para la funcin S
a = 0
c = 5

# Generar los valores de membresa
y = fuzz.smf(x, a, c)

# Graficar
plt.figure()
plt.plot(x, y, 'g', linewidth=1.5, label='S')
plt.title('Funcin de Pertenencia S')
plt.legend()
plt.show()
```

**7. Funcin de Pertenencia Z (zmf)**

La funcin Z es el complemento de la funcin S y tambin utiliza dos parmetros: a y c.

```
# Parmetros para la funcin Z
a = -5
c = 0

# Generar los valores de membresa
y = fuzz.zmf(x, a, c)

# Graficar
plt.figure()
plt.plot(x, y, 'r', linewidth=1.5, label='Z')
plt.title('Funcin de Pertenencia Z')
plt.legend()
plt.show()
```
