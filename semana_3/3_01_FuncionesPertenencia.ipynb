# SciKit-Fuzzy: Funciones de Pertenencia

En SciKit-Fuzzy, puedes implementar las siete funciones de pertenencia más comúnmente usadas en la lógica difusa. Estas funciones son Triangular, Trapezoidal, Gaussiana, Sigmoidal, Bell, S, Z.

```
!pip install scikit-fuzzy
```

```
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

**1. Función de Pertenencia Triangular (trimf)**

La función triangular se define por tres puntos: el inicio, el pico y el final.

```
x = np.linspace(0, 40, 1000)
mf_triangular = fuzz.trimf(x, [10, 20, 30])

plt.figure()
plt.plot(x, mf_triangular, 'b', linewidth=1.5, label='Triangular')
plt.title('Función de Pertenencia Triangular')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**2. Función de Pertenencia Trapezoidal (trapmf)**

La función trapezoidal se define por cuatro puntos: el inicio de la rampa ascendente, el inicio de la meseta, el final de la meseta y el inicio de la rampa descendente.

```
mf_trapezoidal = fuzz.trapmf(x, [10, 15, 25, 30])

plt.figure()
plt.plot(x, mf_trapezoidal, 'r', linewidth=1.5, label='Trapezoidal')
plt.title('Función de Pertenencia Trapezoidal')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**3. Función de Pertenencia Gaussiana**

La Gaussiana se caracteriza por su centro y desviación estándar.

```
mf_gaussiana = fuzz.gaussmf(x, 20, 5)

plt.figure()
plt.plot(x, mf_gaussiana, 'g', linewidth=1.5, label='Gaussiana')
plt.title('Función de Pertenencia Gaussiana')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**4. Función de Pertenencia Sigmoidal**

La Sigmoidal se define por su pendiente (0.1) y el valor en el centro de la curva S (20).

```
mf_sigmoidal = fuzz.sigmf(x, 20, 1)

plt.figure()
plt.plot(x, mf_sigmoidal, 'k', linewidth=1.5, label='Sigmoidal')
plt.title('Función de Pertenencia Sigmoidal')
plt.xlabel('Temperatura')
plt.ylabel('Grado de pertenencia')
plt.legend()
plt.show()
```

**5. Función de Pertenencia Bell (gbellmf)**

La función de Bell se define mediante tres parámetros: a, b, y c, donde a es el ancho, b es la pendiente y c es el centro de la curva.

```
# Definir el universo de discurso
x = np.linspace(-10, 10, 100)

# Parámetros para la función Bell
a = 2
b = 3
c = 0

# Generar los valores de membresía
y = fuzz.gbellmf(x, a, b, c)

# Graficar
plt.figure()
plt.plot(x, y, 'b', linewidth=1.5, label='Bell')
plt.title('Función de Pertenencia Bell')
plt.legend()
plt.show()
```

**6. Función de Pertenencia S (smf)**

La función S es útil para modelar transiciones suaves y se define por dos parámetros: a y c, que son los puntos de inflexión de la curva.

```
# Parámetros para la función S
a = 0
c = 5

# Generar los valores de membresía
y = fuzz.smf(x, a, c)

# Graficar
plt.figure()
plt.plot(x, y, 'g', linewidth=1.5, label='S')
plt.title('Función de Pertenencia S')
plt.legend()
plt.show()
```

**7. Función de Pertenencia Z (zmf)**

La función Z es el complemento de la función S y también utiliza dos parámetros: a y c.

```
# Parámetros para la función Z
a = -5
c = 0

# Generar los valores de membresía
y = fuzz.zmf(x, a, c)

# Graficar
plt.figure()
plt.plot(x, y, 'r', linewidth=1.5, label='Z')
plt.title('Función de Pertenencia Z')
plt.legend()
plt.show()
```
